We would like to create a deciding factor column for each student. We standardize several columns and then pick the most important factor from a lookup table. If the standardized value is above 0.8, the student will be accepted.

We will start by creating a standardized column for `CGPA`, `GRE Score` and `LOR`. We will name these columns `CGPA_std`, `GRE_std`, and `LOR_std` respecively.

Recall that standardizing a column is done by subtracting the mean of the column from all observations in the column and then dividing each observation in the column by the column's standard deviation.

In the cell below, write a standardization function (a function that takes a column as input and returns a standardized column as output). Make sure to use the `numpy` versions of mean and standard deviation.

```python
def standardize(col):
    """
    This function takes a column from a dataframe 
    and returns a standardized column by subtracting the column's mean
    and dividing by the column's standard deviation.
    """
    mean = np.mean(col)
    std_dev = np.std(col)
    standardized_col = (col - mean) / std_dev
    return standardized_col

# Apply the standardization function to the specified columns and create new standardized columns
admission_data['CGPA_std'] = standardize(admission_data.index.get_level_values('CGPA'))
admission_data['GRE_std'] = standardize(admission_data.index.get_level_values('GRE Score'))
admission_data['LOR_std'] = standardize(admission_data['LOR'])

# Display the first few rows to confirm the new columns
admission_data.head()
```

Now create the standardized columns `CGPA_std`, `GRE_std`, and `LOR_std` and add them to the `admissions` dataframe.

```python
admission_data.head()
```

We will generate the decision choice at random using the code below. Please run the cell.

```python
# Libraries
from random import choices
```

```python
# Standardized columns
std_columns = ['CGPA_std', 'GRE_std', 'LOR_std']

# Generate the decision choice at random for each row
decision_choice = choices(std_columns, k=admission_data.shape[0])

# Add the decision_choice to the dataframe
admission_data['decision_choice'] = decision_choice

# Display the first few rows to confirm the new column
admission_data.head()
```

Now create the deciding column using the `lookup` function. The lookup column is `decision_choice` found above. Call the column resulting from the lookup function `deciding_column` and add it to the `admissions` dataframe.

```python
# Create the deciding_column using the lookup function
admission_data['deciding_column'] = admission_data.lookup(row_labels=admission_data.index, col_labels=admission_data['decision_choice'])

# Display the first few rows to confirm the new column
admission_data.head()
```

```python
#Since the lookup function is deprecated and not available in newer versions of pandas, we can use an alternative approach with the apply function to achieve the same result.
#Function to fetch the value from the appropriate standardized column

def get_decision(row):
    return row[row['decision_choice']]

# Apply the function to each row
admission_data['deciding_column'] = admission_data.apply(get_decision, axis=1)

# Display the first few rows to confirm the new column
admission_data.head()
```

Create a column called `decision` in the `admissions` dataframe. Assign 1 to this column if the value of `deciding_column` is greater than 0.8 and 0 otherwise.

```python
# Create the decision column
admission_data['decision'] = np.where(admission_data['deciding_column'] > 0.8, 1, 0)

# Display the first few rows to confirm the new column
admission_data.head()
```

How many applicants will be accepted to the program using the decision column? Compute the result below.

```python
# Compute the number of applicants accepted to the program
num_accepted = admission_data['decision'].sum()

# Display the result
print(f"The number of applicants accepted to the program is: {num_accepted}")
```

# Challenge 3 - Method Chaining

To increase our coding efficiency, let's make a number of changes to our dataframe in one line of code.

In the cell below, remove all non character symbols from the column names, replace all spaces in column names to underscores, and change all upper case characters to lower case. Assign these new column values to `admissions.columns`.

```python
# Method chaining to remove non-character symbols, replace spaces with underscores, and change to lower case
admission_data.columns = admission_data.columns.str.replace('[^A-Za-z]','', regex=True).str.replace(' ','_').str.lower()

# Display the updated columns to confirm the changes
admission_data.head()
```
